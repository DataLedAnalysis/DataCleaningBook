```python
import pandas as pd
import openpyxl as xl


pth = r"C:\Users\wasayabd\OneDrive - Mars Inc\Documents\DataSets\Film_Awards.xlsx"
df = pd.read_excel(pth,header=1) #1 8 15 = 7 incremented
df['Year'] = df['Year'].astype(pd.StringDtype())

lst = ['Academy Award','Golden Globe - Drama','Golden Globe - Musical or Comedy']

df.dropna(subset=['Year','Academy Award','Golden Globe - Drama','Golden Globe - Musical or Comedy'], inplace=True)

patternToDel = '\d{4}'

df = df[df['Year'].str.match(patternToDel) == True] #we have the dataframe now

df


```

```python
#calculating rows
r = pd.read_excel(pth,header=None) #1 8 15 = 7 incremented
rowsNum = len(r.index) + 1 #not inclusive

wb = xl.load_workbook(pth)
sheet = wb.worksheets[0]
def getAttribute(n):
    cellValue = sheet['A' + str(n)].value
    ind = n -1
    return ind,cellValue

myDict = {'row_number':[],'attribute':[]}

for Val in range(1,rowsNum,7):
    attr = getAttribute(Val)
    myDict['row_number'].append(attr[0])
    myDict['attribute'].append(attr[1])
print(myDict)

```
